```python
#converting that string to integers and floats
price=[]
for i in y["price"]:
    price.append(int(i))
print(price,end="")

peak_rpm=[]
for i in y["peak-rpm"]:
    peak_rpm.append(int(i))
print(peak_rpm,end="")

norm=[]
for i in y['normalized-losses']:
    norm.append(int(i))
print(norm,end="")

nod=[]
for i in y['num-of-doors']:
    nod.append(int(i))
print(nod,end="")



bore=[]
for i in y['bore']:
    bore.append(float(i))
print(bore,end="")

stroke=[]
for i in y['stroke']:
    stroke.append(float(i))
print(stroke,end="")

hp=[]
for i in y['horsepower']:
    hp.append(int(i))
print(hp,end="")
```

```python
#assigning that converted integers to the columns
y['horsepower']=hp

y['stroke']=stroke

y['bore']=bore

y['num-of-doors']=nod

y['normalized-losses']=norm

y["peak-rpm"]=peak_rpm

y["price"]=price
#to avoid the warnings while executing
import warnings
import numpy as np
warnings.filterwarnings("ignore")
#the given car prices are at high values if we plot them then accuracy will not get more 
#for that reason i use log to decrease their values 
#to the lower vales with out changing the values
y["CAR_PRICE"]=np.log(y["price"])

```

```python
#There are some strings which visible to us same as string
#we have to convert them into integer then only we can continue further process
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in y.columns:
    y[i]=le.fit_transform(y[i])
    
#again checking for data type for better conformation
for i in y.columns:
    print(i,y[i].dtype)
```

```python
#as we can see above we taken log for the price and assigned it to new column. so no need with
#column old one, drop it then.

car=y.drop(["price"],axis=1)
car
```

```python
#Now seperating the data as input and output data
a=car.drop(["CAR_PRICE"],axis=1)
y=car["CAR_PRICE"]

#To get better accuracy i wannna standardize the data
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
for i in a.columns:
    a[i]=scale.fit_transform(a[[i]])
    
a
```

```python
#i saw one error at that time of fitting to resolve that i used two digit data.
#error is float64 is heavy cant process with that data
x=round(a,2)
#splitting the data into training[70%] and testing[30%] 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python

from sklearn.linear_model import LinearRegression
#INITIALIZING THE ALGORITHM BEFORE FITTING
lr=LinearRegression()
#fit the data to train the model
model=lr.fit(x_train,y_train)
#predict the patterns which are known and which are unknown
y_predict=model.predict(x_test)
#taking the score[accuracy] of our trained model
from sklearn.metrics import r2_score
round((r2_score(y_test,y_predict))*100,2)
```

```python
#plotting the values with actual and predicted
fig = plt.figure(figsize = (10,8))
plt.scatter(y_test,y_predict, edgecolor = 'black')
fig.suptitle('y_test vs y_pred', fontsize=20)               
plt.xlabel('y_test', fontsize=18)                          
plt.ylabel('y_pred', fontsize=18)
```

```python
#taking index of the y_predicted and put it on x_axis 
c = [i for i in range(1,63,1)]
#comparing the actual and predicted values 
fig = plt.figure(figsize=(10,8))
plt.plot(c,y_test, color='black', linewidth=3, linestyle="--")
plt.plot(c,y_predict, color='red',  linewidth=3, linestyle="--")
fig.suptitle('Comparing Actual and Predicted', fontsize=20)  
plt.xlabel('Index', fontsize=18)                            
plt.ylabel('Views', fontsize=18)
```

```python
#representing the errors in the predicted data set with DISTRIBUTION PLOT
fig = plt.figure(figsize=(10,8))
sns.distplot((y_test - y_predict), bins = 20, color = 'maroon')
fig.suptitle('Error Terms', fontsize = 20)                   
plt.xlabel('Errors', fontsize = 18)
```
