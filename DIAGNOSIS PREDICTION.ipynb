# DIAGNOSIS PREDICTION BY KNN ALGORITHM

```python

#IMPORTING DATASET
import pandas as pd
df=pd.read_csv("E:\DSTUTORIALS\Dataset-c.csv")
print("DATASET",df)

#DEFINING TARGETING FEATURE
y=df["diagnosis"]
print("TARGET FEATURES",y)
#DEFINING INPUT FEATURES
x=df.drop(["diagnosis"],axis=1)
print("INPUT FEATURES",x)
#AVOIDING WARNING IF ANY COME IN THE MIDDLE OF PROGRAMME
import warnings
warnings.filterwarnings("ignore")

#SPLITTING THE DATA INTO TRAINNG DATA AND TESTING DATA
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)

#SPLITTING THE WHOLE DATA INTO WHOLE DATA FOR GET MORE ACCURACY
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

#I WANT COMPARE THE ACCURACY LIKE WHICH ALGORITHM GIVES MORE ACCURACY
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

#CREATE A LIST TO GATHER MODEL NAME AND INSTANCE OF MODEL 
algname=[]

#TAKE SEPERATELY MODEL NAME AND INSTANCE OF MODEL AND NEXT APPEND THEM TO ABOVE EMPTY LIST
compKNN=["KNN",KNeighborsClassifier()]
compSVC=["SVC",SVC()]
algname.append(compKNN)
algname.append(compSVC)


#TAKE TWO MORE LISTS TO COLLECT NAME OF MODEL AND THE MODEL RESULT
names=[]
result=[]
for name, model in algname:
    kfold=KFold(n_splits=10)
    cv_result=cross_val_score(model,x_train,y_train,cv=kfold,scoring="accuracy")
    result.append(cv_result.mean())
    names.append(name)
    print("MODEL NAME",names,"RESULTS OF MODEL",result)
#I FINDED THAT KNN ALGORITHM GIVING MORE ACCURACY SO THAT MY ALGORITHM IS PROCESSING WITH KNN 
knb=KNeighborsClassifier()
model=knb.fit(x_train,y_train)
y_predict=model.predict(x_test)
print("PREDICTED VALUES",y_predict)

#DEFINE CONFUSION MATRIX TO COMPARE ACTUAL AND PREDICTED VALUES
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
print("CONFUSION MATRIX",cm)

#CLASSIFICATION REPORT IS USED TO SHOW THE PRECISION,RECALL,F1-SCORE,SUPPORT
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_predict)
print("CLASSIFICATION REPORT",cr)


#model.score(x_test,y_test)*100

#AGAIN SEE THIS MODEL ACCURACY BY THIS OR ABOVE COMMENTED ONE [EITHER ONE]
from sklearn.metrics import accuracy_score
m=accuracy_score(y_test,y_predict)*100
print("ACCURACY OF MODEL",m)


#model.predict([[842517,20.57,17.77,132.9,1326,0.08474,0.07864,0.0869,0.07017,0.1812,0.05667,0.5435,
#                0.7339,3.398,74.08,0.005225,0.01308,0.0186,0.0134,0.01389,0.003532,24.99,23.41,158.8,
#                1956,0.1238,0.1866,0.2416,0.186,0.275,0.08902]])

#PREPROCESING: IT IS USED TO CONVERT LABEL DATA TO NUMERICAL DATA
from sklearn.preprocessing import LabelEncoder
m=LabelEncoder()
r=m.fit_transform(y_test)
s=m.fit_transform(y_predict)
#IT INDICATES HOW MUCH PERCENTAGE POSITIVE SAMPLES ARE MORE THAN NEGATIVE SAMPLES
from sklearn.metrics import roc_auc_score
score=roc_auc_score(r,s)
print("RECEIVER_OPERATIONAL_CHARACTERISTIC CURVE &AREA_UNDER_CURVE SCORE",score*100)
```
